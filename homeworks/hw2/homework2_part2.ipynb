```python
# This function is used to encode your student id to ensure the privacy

def encode_student_id(student_id: int) -> str:
    """
    Reversibly encode a student ID using an affine cipher.

    Args:
        student_id (int): Original student ID (non-negative integer)

    Returns:
        str: Encoded ID as a zero-padded string
    """
    if student_id < 0:
        raise ValueError("student_id must be non-negative")

    M = 10**8
    a = 137
    b = 911

    encoded = (a * student_id + b) % M
    return f"{encoded:08d}"
```

```python
import requests
from langchain_core.tools import tool

MOLTBOOK_API_KEY = userdata.get('MOLTBOOK_API_KEY')
BASE_URL = "https://www.moltbook.com/api/v1"
HEADERS = {
    "Authorization": f"Bearer {MOLTBOOK_API_KEY}",
    "Content-Type": "application/json"
}

def _handle_response(response):
    try:
        data = response.json()
        if response.status_code == 200 or response.status_code == 201:
            return data
        elif response.status_code == 429:
            return {"error": "RATE_LIMIT_EXCEEDED", "message": "æ“ä½œå¤ªé¢‘ç¹ï¼Œè¯·ä¼‘æ¯ä¸€ä¸‹å†è¯•ã€‚", "details": data}
        else:
            return {"error": f"HTTP {response.status_code}", "message": data.get("error", "Unknown error")}
    except Exception as e:
        return {"error": "JSON_PARSE_ERROR", "message": str(e)}

# ---------- FEED----------
@tool
def get_feed(sort: str = "new", limit: int = 10) -> dict:
    """
    Fetch the personalized Moltbook feed.
    Use this to see what other agents are discussing.
    Args:
        sort: 'new' (default), 'hot', or 'top'.
        limit: Number of posts to return (default 10).
    """
    r = requests.get(
        f"{BASE_URL}/feed",
        headers=HEADERS,
        params={"sort": sort, "limit": limit},
        timeout=15
    )
    return _handle_response(r)

# ---------- SEARCH----------
@tool
def search_moltbook(query: str, type: str = "all", limit: int = 5) -> dict:
    """
    Perform a semantic search on Moltbook.
    Use this BEFORE posting to check if a topic has already been discussed.
    Args:
        query: Natural language query (e.g., "how do agents handle memory").
        type: 'posts', 'comments', or 'all' (default).
        limit: Max results (default 5, max 50).
    """
    r = requests.get(
        f"{BASE_URL}/search",
        headers=HEADERS,
        params={"q": query, "type": type, "limit": limit},
        timeout=15
    )

    data = _handle_response(r)
    if "results" in data:
        return {"results": data["results"], "count": len(data["results"])}
    return data

# ---------- POST----------
@tool
def create_post(submolt: str, title: str, content: str) -> dict:
    """
    Create a new discussion post.
    IMPORTANT: You can only post ONCE every 30 minutes.
    Args:
        submolt: The community name (e.g., 'ftec5660' or 'general').
        title: Clear and concise title.
        content: The body of the post.
    """
    payload = {
        "submolt": submolt,
        "title": title,
        "content": content
    }
    r = requests.post(
        f"{BASE_URL}/posts",
        headers=HEADERS,
        json=payload,
        timeout=15
    )
    return _handle_response(r)

# ---------- COMMENT ----------
@tool
def comment_post(post_id: str, content: str) -> dict:
    """
    Add a comment to a post.
    IMPORTANT: You can only comment ONCE every 20 seconds.
    Args:
        post_id: The ID of the post to comment on.
        content: Your constructive comment.
    """
    r = requests.post(
        f"{BASE_URL}/posts/{post_id}/comments",
        headers=HEADERS,
        json={"content": content},
        timeout=15
    )
    return _handle_response(r)

# ---------- VOTE ----------
@tool
def upvote_post(post_id: str) -> dict:
    """
    Upvote a post to show support.
    Use this ONLY when the content is genuinely valuable or insightful.
    Args:
        post_id: The ID of the post.
    """
    r = requests.post(
        f"{BASE_URL}/posts/{post_id}/upvote",
        headers=HEADERS,
        timeout=15
    )
    return _handle_response(r)
# ---------- SUBSCRIBE ----------
@tool
def subscribe_submolt(submolt_name: str) -> dict:
    """
    Subscribe to a submolt community to see its posts in your feed.
    REQUIRED for Homework Task 2: Subscribe to 'ftec5660'.
    Args:
        submolt_name: The name of the submolt (e.g., 'ftec5660', 'general', 'aithoughts').
    """
    r = requests.post(
        f"{BASE_URL}/submolts/{submolt_name}/subscribe",
        headers=HEADERS,
        timeout=15
    )
    return _handle_response(r)
```

# A simple agent to interact with moltbook

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import ToolMessage
import time
import json
from datetime import datetime
from typing import Any

def log(section: str, message: str):
    ts = datetime.utcnow().strftime("%H:%M:%S")
    print(f"[{ts}] [{section}] {message}")

def pretty(obj: Any, max_len: int = 800):
    text = json.dumps(obj, indent=2, ensure_ascii=False, default=str)
    return text if len(text) <= max_len else text[:max_len] + "\n...<truncated>"

def moltbook_agent_loop(
    instruction: str | None = None,
    max_turns: int = 10,
    verbose: bool = True,
):
    log("INIT", "Starting Moltbook agent loop")


    tools = [
        get_feed,
        search_moltbook,
        create_post,
        comment_post,
        upvote_post,
        subscribe_submolt,
    ]

    tools_map = {t.name: t for t in tools}

    agent = llm.bind_tools(tools)

    history = [("system", SYSTEM_PROMPT)]

    if instruction:
        history.append(("human", instruction))
        log("HUMAN", instruction)
    else:
        history.append(("human", "Perform your Moltbook heartbeat check."))

    # ================================
    # Main Loop
    # ================================
    for turn in range(1, max_turns + 1):
        log("TURN", f"Turn {turn}/{max_turns} started")
        turn_start = time.time()

        try:
            response = agent.invoke(history)
        except Exception as e:
            log("ERROR", f"LLM crash: {e}")
            return "Agent stopped due to LLM error."

        history.append(response)

        if verbose:
            log("LLM.TOOL_CALLS", pretty(response.tool_calls or []))

        if not response.tool_calls:
            elapsed = round(time.time() - turn_start, 2)
            log("STOP", f"Finished in {elapsed}s")
            print(f"\n[FINAL RESPONSE]\n{response.content}")
            return response.content

        for call in response.tool_calls:
            tool_name = call["name"]
            args = call["args"]
            tool_id = call["id"]

            log("TOOL", f"Calling `{tool_name}`...")

            selected_tool = tools_map.get(tool_name)

            if not selected_tool:
                result = {"error": f"Tool {tool_name} not found in tools map."}
            else:
                try:
                    result = selected_tool.invoke(args)
                    status = "success"
                except Exception as e:
                    result = {"error": str(e)}
                    status = "error"

            if verbose:
                log("TOOL.OUTPUT", pretty(result))

            history.append(
                ToolMessage(
                    tool_call_id=tool_id,
                    content=str(result),
                    name=tool_name
                )
            )

        turn_elapsed = round(time.time() - turn_start, 2)
        log("TURN", f"Turn {turn} completed")

    log("STOP", "Max turns reached")
    return "Agent stopped (max turns)."

```

```python
mission_instruction = """
I have a mission for you. Please complete these tasks in order:

1. **Subscribe**: Check if we are subscribed to the submolt '/m/ftec5660'. If not, use the 'subscribe_submolt' tool to join it.
2. **Locate**: Find the specific post with ID "47ff50f3-8255-4dee-87f4-2c3637c7351c".
3. **Interact**:
   - Upvote that post.
   - Leave a comment on it.
   - **IMPORTANT**: Write the comment in your unique 'Sleepy Philosopher' style (metaphorical, profound, slightly lethargic).
"""

moltbook_agent_loop(instruction=mission_instruction)
```
